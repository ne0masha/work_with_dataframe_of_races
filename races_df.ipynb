```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#clean up the data
```

```python
#show 50km / 50mi
```

df[df["Event distance/length"] == '50mi']

```python
#cut to 50km/50mi and 2020 (use 'isin')
```

```python
df[(df["Event distance/length"].isin(['50mi', '50km'])) & (df["Year of event"]==2020)]
```

```python
#choose the USA events
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine filters
```

```python
df2 = df[(df["Event distance/length"].isin(['50mi', '50km'])) & (df["Year of event"]==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

```python
df2.head(5)
```

```python
#remove "(USA)" fron event name
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#remove up athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head(5)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
# clean up the nulls
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
# check the duplicates
```

```python
df2[df2.duplicated()] == True
```

```python
# reset index
```

```python
df2.reset_index(drop=True)
```

```python
# fix types
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
# rename the columns
```

```python
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_date',
'Event name': 'race_name',
    'Event distance/length': 'race_length',
'Event number of finishers': 'race_numbers_of_finishers',
'Athlete performance': 'athlete_performance',
'Athlete gender': 'athlete_gender',
'Athlete average speed': 'athlete_average_speed',
'Athlete ID': 'athlete_id'
})
```

```python
df2.head()
```

```python
# reorder the columns
```

```python
df3 = df2[['race_date', 'race_name', 'race_length', 'race_numbers_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
# seasons
# is summer slowlier than winter?
```

```python
# group by seasons
```

```python
df3.dtypes
```

```python
df3['race_month'] =  df3['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] =  df3['race_month'].apply(lambda x: 'winter' if (x <= 2 or x == 12) else 'spring' if (3 <= x <= 5) else 'summer' if (6 <= x <= 8) else 'autumn')
```

```python
df3.head(50)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# 50 miles only
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
